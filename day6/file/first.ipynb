# q2 dictionary Merger with Aggregation.

```python
inv1={
    "apples": 10,
    "bananas":5
}
inv2={
    "bananas":20,
    "oranges":15
}

def merge_inventories(inv1,inv2):
    merge = inv1.copy()
    for key,value in inv2.items():
        if key in merge:
            merge[key] += value
        else:
            merge[key] = value
    return merge
asd = merge_inventories(inv1,inv2)
print(asd)
```

# Q3 string Compress


```python

text = "aabccccaaa"
def compress_string(text):
    visited = {}
    for i in text:
        if i in visited:
            visited[i]+=1
        else:
            visited[i]=1
    return visited
print(compress_string(text))
```
